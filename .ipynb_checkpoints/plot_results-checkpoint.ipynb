```python
##### entropy plots
from scipy.stats import entropy

results_dir = "results/distributions/UUP/enum_"
df_non_condorcet = pd.DataFrame(columns=['entropy', 'n_condorcet'])
for i in range(196):
    uup_dir = results_dir + str(i) + "/m_3/"
    uup_condorcet = uup_dir + "condorcet.csv"
    uup_prob_vector = uup_dir + "prob_vector.csv"
    prob = pd.read_csv(uup_prob_vector, index_col=0)
    #print(list(prob["0"]))
    ent = entropy(list(prob["0"]), base=2)
    condorcet_df = pd.read_csv(uup_condorcet)
    proportion = condorcet_df.loc[0]['n_condorcet']/10000
    df_non_condorcet.loc[len(df_non_condorcet)] = [ent, proportion]

uniform_condorcet = "results/distributions/uniform/m_3/condorcet.csv"
ent = entropy([1/6]*6, base=2)
condorcet_df = pd.read_csv(uniform_condorcet)
proportion = condorcet_df.loc[0]['n_condorcet']/10000
df_non_condorcet.loc[len(df_non_condorcet)] = [ent, proportion]
plt.scatter(df_non_condorcet['entropy'], df_non_condorcet['n_condorcet'])
plt.xlabel("Entropy (base 2)")
plt.ylabel("Proportion of Condorcet Winners")

plt.title("Proportion of Condorcet Winners vs Entropy, 3 candidates")
plt.savefig("plots/entropy_ncondorcet.png")
plt.show()
```
